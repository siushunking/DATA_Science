```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
X = pd.DataFrame(X)
sex = dataset.iloc[:,5].values
country = dataset.iloc[:,4].values
```

```python
sex = pd.get_dummies(sex)
sex = sex.iloc[:,0:1]
country = pd.get_dummies(country)
country = country.iloc[:,0:2]
```

```python
x1 = X.iloc[:,3:9]
```

```python
x2 = X.iloc[:,0:1]
```

```python
sex_country = pd.concat([sex, country], axis=1)
```

```python
X = pd.concat([x1, x2, sex_country], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
   tf.keras.layers.Dense(units=100, activation=tf.nn.relu, input_dim=10),
   tf.keras.layers.Dense(units=100, activation=tf.nn.relu),
   tf.keras.layers.Dense(units=2, activation=tf.nn.softmax)                                 
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras import backend as K
```

```python
X_train = K.cast_to_floatx(X_train)
y_train1 = K.cast_to_floatx(y_train)
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=128)
```

```python
score = model.evaluate(X_train, y_train, batch_size=128)
```

```python
predict = model.predict(X_train)
```
